```
import numpy as np

class MyMultiLR:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None

    def fit(self, X_train, y_train):
        # इनपुट डेटा को सुरक्षित रूप से NumPy array में बदलना
        X_train = np.asarray(X_train)
        y_train = np.asarray(y_train)

        # इंटरसेप्ट के लिए पहले कॉलम (index 0) में 1 (ones) जोड़ना
        X_train_new = np.insert(X_train, 0, 1, axis=1)

        # नॉर्मल इक्वेशन फ़ॉर्मूला: (X^T * X)^-1 * X^T * Y
        betas = np.linalg.inv(np.dot(X_train_new.T, X_train_new)).dot(X_train_new.T).dot(y_train)

        # इंटरसेप्ट और कोएफिशिएंट्स को अलग करके स्टोर करना
        self.intercept_ = betas[0]
        self.coef_ = betas[1:]

    def predict(self, X_test):
        # टेस्ट डेटा को NumPy array में बदलना
        X_test = np.asarray(X_test)

        # y_pred = X * coef + intercept फ़ॉर्मूला लागू करना
        return np.dot(X_test, self.coef_) + self.intercept_
```

```
mylr = MyMultiLR()
```
